```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from math import sqrt
```

# preprocessing

```python
df = pd.read_csv('C:/Users/Lenovo/Desktop/selenium/selenium/EV/output.csv',encoding='gbk')

df['price'] = df['price'].str[1:-1].astype(float)
df['price'] = df['price']*10000

df['new_car_price'] = df['new_car_price'].str[5:-5].astype(float)
df['new_car_price'] = df['new_car_price']*10000

df['miles'] = df['miles'].str[:-3].astype(float)
df['miles'] = df['miles']*10000

df['RV_ratio'] = df['price']/df['new_car_price']

df['record_date'] = pd.to_datetime(df['record_date'])
df['licensing_time'] = pd.to_datetime(df['licensing_time'])

df['months'] = ((df['record_date'] - df['licensing_time'])/np.timedelta64(1, 'M'))
df['months'] = df['months'].astype(int)

df['battery_volume'] = df['battery_volume'].replace('-', np.nan)
df['battery_power_data'] = df['battery_power_data'].replace('-', np.nan)
df['battery_torque_data'] = df['battery_torque_data'].replace('-', np.nan)
df['endurance_miles'] = df['endurance_miles'].replace('-', np.nan)

df['battery_volume'] = df['battery_volume'].astype(float)
df['battery_power_data'] = df['battery_power_data'].astype(float)
df['battery_torque_data'] = df['battery_torque_data'].astype(float)
df['endurance_miles'] = df['endurance_miles'].astype(float)

#df['car_number'] = df['car_number'].str[4:15]

df
```

```python
df = df.drop_duplicates(subset=['car_number'],keep='first')
```
