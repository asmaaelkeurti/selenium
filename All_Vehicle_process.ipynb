## Processing

```python
def sale_price(string):
    try:
        first_d = re.search("\d",string).start()
        first_D = first_d + re.search("万",string[first_d:]).start()
        second_d = first_D + re.search("\d",string[first_D:]).start()
        second_D = second_d + re.search("万",string[second_d:]).start()
        return string[first_d:first_D]
    except AttributeError:
        print(string)
        return np.NaN
    
def MSRP_with_Tax(string):
    try:
        first_d = re.search("\d",string).start()
        first_D = first_d + re.search("万",string[first_d:]).start()
        second_d = first_D + re.search("\d",string[first_D:]).start()
        second_D = second_d + re.search("万",string[second_d:]).start()
        return string[second_d:second_D]
    except AttributeError:
        print(string)
        return np.NaN
```

```python
df = pd.read_csv('C:/Users/Lenovo/Desktop/output.csv',encoding='gbk')

#删除 补贴
df = df[~df['info'].str.contains('补贴')]
# Split
df['info'] = df['info'].str.split('|')
df['model'] = df['info'].apply(lambda x:x[0])

df['year'] = df['model'].str[-5:]
df['model'] = df['model'].str[:-5]

df['miles'] = df['info'].apply(lambda x:x[1])
df['info3'] = df['info'].apply(lambda x:x[2])


df['sale_price'] = df['info3'].apply(lambda x:sale_price(x))
df['MSRP_with_Tax'] = df['info3'].apply(lambda x:MSRP_with_Tax(x))
df['location'] = df['info3'].apply(lambda x:x[:re.search("\d",x).start()])

df
```
