```python
def sale_price(string):
    try:
        first_d = re.search("\d",string).start()
        first_D = first_d + re.search("万",string[first_d:]).start()
        second_d = first_D + re.search("\d",string[first_D:]).start()
        second_D = second_d + re.search("万",string[second_d:]).start()
        return string[first_d:first_D]
    except AttributeError:
        return np.NaN
    
def MSRP_with_Tax(string):
    try:
        first_d = re.search("\d",string).start()
        first_D = first_d + re.search("万",string[first_d:]).start()
        second_d = first_D + re.search("\d",string[first_D:]).start()
        second_D = second_d + re.search("万",string[second_d:]).start()
        return string[second_d:second_D]
    except AttributeError:
        return np.NaN
```

```python
def process_output(file_location):
    df = pd.read_csv(file_location,encoding='gbk')

    #删除 补贴
    df = df[~df['info'].str.contains('补贴')]
    # Split
    df['info'] = df['info'].str.split('|')
    df['model'] = df['info'].apply(lambda x:x[0])

    df['year'] = df['model'].str[-5:]
    df['model'] = df['model'].str[:-5]

    df['miles'] = df['info'].apply(lambda x:x[1])
    df['info3'] = df['info'].apply(lambda x:x[2])


    df['sale_price'] = df['info3'].apply(lambda x:sale_price(x)).astype(float)
    df['MSRP_with_Tax'] = df['info3'].apply(lambda x:MSRP_with_Tax(x)).astype(float)
    df['location'] = df['info3'].apply(lambda x:x[:re.search("\d",x).start()])
    
    df['vehicle_id'] = df['vehicle_id'].apply(lambda x:x[:re.search('@',x).start()])
    
    df = df[['model','year','miles','sale_price','MSRP_with_Tax','location','vehicle_type','record_date','vehicle_id','link']]
    df['last_update_date'] = df['record_date']
    
    df = df.drop_duplicates(subset=['model','vehicle_id','location'],keep='first')
    
    return df
```

```python
def append_new_output(new_df,old_df):
    new_output = new_df
    old_output = old_df
    
    update_df = old_output.drop(['link','last_update_date'],axis=1).\
            merge(new_output[['vehicle_id','location','model','last_update_date','link']],
            on=['vehicle_id','location','model'],how='inner')
        
    left_join_non_update = old_output.merge(new_output[['vehicle_id','location','model','last_update_date']].rename(columns={'last_update_date':'x'}),
                               on=['vehicle_id','location','model'],how='left')
    non_update_df = left_join_non_update[left_join_non_update['x'].isnull()].drop(['x'],axis=1)
    
    left_join = new_output.merge(old_output[['vehicle_id','location','model','last_update_date']].rename(columns={'last_update_date':'x'}),on=['vehicle_id','location','model'],how='left')
    new_add_df = left_join[left_join['x'].isnull()].drop(['x'],axis=1)
    
    appended_df = update_df.append([non_update_df,new_add_df])
    
    return appended_df
```
